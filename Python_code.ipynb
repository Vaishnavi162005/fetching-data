```python
import time
import pandas as pd
import requests
import gspread
from oauth2client.service_account import ServiceAccountCredentials

# Fetching of data from CoinGecko API
url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=50"

while True:
    response = requests.get(url)
    data = response.json()
    
    # Extraction of required fields
    crypto_data = [
        {
            "Name": coin["name"],
            "Symbol": coin["symbol"],
            "Current Price (USD)": coin["current_price"],
            "Market Cap (USD)": coin["market_cap"],
            "24h Trading Volume (USD)": coin["total_volume"],
            "24h Price Change (%)": coin["price_change_percentage_24h"]
        }
        for coin in data
    ]
    
    df = pd.DataFrame(crypto_data)
    print(df)
    
    # Google Sheets Setup
    SPREADSHEET_URL = "https://docs.google.com/spreadsheets/d/1xX_JXttvGt_JCY9uJuk1unnh6ygX1Febs_L8f5ytafA/edit?gid=0#gid=0"
    SHEET_NAME = "Sheet1" 
    
    # Authentication n connection with Google Sheets
    def connect_to_google_sheets():
        scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
        creds = ServiceAccountCredentials.from_json_keyfile_name("C:/Users/saksh/Downloads/live-cryptocurrency-data-375c1a785b3c.json", scope)
        client = gspread.authorize(creds)
        return client.open_by_url(SPREADSHEET_URL).worksheet(SHEET_NAME)
    
    sheet = connect_to_google_sheets()
    sheet.clear()
    # Append rows from DataFrame to Google Sheets
    sheet.append_row(df.columns.tolist())  # Add headers (optional)
    for _, row in df.iterrows():
        sheet.append_row(row.tolist())  # Append each row as a list

        
    print("\nGoogle Sheet updated successfully.")
            
    # Top 5 cryptocurrencies by market cap
    top_5 = df.nlargest(5, "Market Cap (USD)")
    print("\nTop 5 Cryptocurrencies by Market Cap:")
    print(top_5[["Name", "Market Cap (USD)"]])
    
    # Average price of top 50 cryptocurrencies
    average_price = df["Current Price (USD)"].mean()
    print(f"\nAverage Price of Top 50 Cryptocurrencies: ${average_price:.2f}")
            
    # Highest 24-hour percentage change
    highest_change = df.nlargest(1, "24h Price Change (%)")
    print("\nHighest 24-Hour Percentage Change:")
    print(highest_change[["Name", "24h Price Change (%)"]])
    
    # lowest 24-hour percentage change
    lowest_change = df.nsmallest(1, "24h Price Change (%)")
    print("\nLowest 24-Hour Percentage Change:")
    print(lowest_change[["Name", "24h Price Change (%)"]])
    
    
    time.sleep(300)
```
